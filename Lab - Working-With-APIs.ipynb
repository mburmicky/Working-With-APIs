**API 1) SPORTMONKS:** getting information about football fixtures

```python
import requests
import getpass
import json
apikey = getpass.getpass()
```

```python
#https://api.sportmonks.com/v3/football/fixtures?api_token=YOUR_TOKEN
```

```python
response = requests.get('https://api.sportmonks.com/v3/football/fixtures?api_token='+apikey+'&include=statistics')
response
results = response.json()
results
```

```python
#This is a list of dictionaries
results['data']
```

```python
import pandas as pd
fixtures = pd.DataFrame(results["data"])
fixtures
```
